```python
sys = RK_own.system(1/50,1,2)
p_in = np.array([1,1])
z_in = np.array([2.04,9.06])
st = RK_own.state(sys,p_in,z_in)
```

```python
num_steps = 1000
tmax = 30
res = st.runge_kutta4(tmax,num_steps)
p_vals = res[0]
z_vals = res[1]
```

```python
E = st.E_t_tr(p_vals,z_vals)
norm = st.length(p_vals,z_vals)
t = np.linspace(0,tmax,num_steps+1)
```

```python
fig = RK_own.go.Figure()
fig.add_scatter(x = t, y = norm)
fig.update_layout(template= RK_own.fig_template, xaxis = dict(title='t',tickmode = 'array',ticklen = 5),
                  yaxis = dict(title='norm',tickmode = 'array',ticklen = 5) )
```
